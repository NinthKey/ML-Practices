```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
#load csv data files
temp_mpg = []
temp_weight = []
with open('mpg.csv', newline='') as csvfile:
     reader = csv.DictReader(csvfile)
     for row in reader:
         temp_mpg.append(row['mpg (n)'])
         temp_weight.append(row['weight (n)'])
mpg = np.array(temp_mpg).astype(float)
weight = np.array(temp_weight).astype(float)
```

```python
# normalize weight and mpg using mean normalization
'''max_mpg = np.max(mpg)
mean_mpg = np.mean(mpg)
mpg = (mpg - mean_mpg) / max_mpg'''
max_weight = np.max(weight)
mean_weight = np.mean(weight)
weight = (weight - mean_weight) / max_weight
```

```python
# scatter plot showing the input data
plt.scatter(weight,mpg)
plt.show() 
```

```python
theta_0 = 0 # random guess for theta_0
theta_1 = 0 # random guess for theta_1
m = len(mpg) # the number of training samples
epochs = 10000 # number of times of training
alpha = 0.01 # learning rate
cost = np.zeros(epochs) # used to plot cost function
for i in range(epochs):
    sum_t0 = 0
    sum_t1 = 0
    sum_cost = 0
    for j in range(m):
        x_j = float(weight[j])
        y_j = float(mpg[j])
        h = theta_0 + theta_1 * x_j
        sum_t0 = sum_t0 + (h - y_j)
        sum_t1 = sum_t1 + (h - y_j) * x_j
        sum_cost = sum_cost + (h - y_j) **2
    temp_t0 = sum_t0 * 1/m
    temp_t1 = sum_t1 * 1/m
    theta_0 = theta_0 - alpha * temp_t0
    theta_1 = theta_1 - alpha * temp_t1
    cost[i] = 1/(2*m) * sum_cost
```

```python
# plot our results against the orginial data
plt.plot(weight, theta_0 + theta_1 * weight)
plt.scatter(weight,mpg)
plt.show()
```

```python
# plot the cost function
l = range(epochs)
plt.plot(l,cost[l])
plt.show()
```
